```python
import pandas as pd
import numpy as np
base_url = 'https://trisonics-scouting-api.azurewebsites.net/api'


def get_scouted_data_url(secret_key, event_key):
    return f'{base_url}/GetResults?secret_team_key={secret_key}&event_key={event_key}'  # noqa


def get_opr_data_url(secret_key, event_key):
    return f"{base_url}/GetOPRData?secret_team_key={secret_key}&event_key={event_key}"  # noqa


url = get_scouted_data_url('4003data', '2023micmp4')
opr_url = get_opr_data_url('4003data', '2023micmp4')
scouted_data = pd.read_json(url)
opr_data = pd.read_json(opr_url)
```

```python
vardata = scouted_data.drop(['event_key', 'match_key', 'match_notes',
                            'scouter_name', 'secret_team_key'], axis=1)

summed = vardata.groupby('scouting_team').sum().reset_index()
summed.head()
```

```python
demeaned = summed.copy()
numerical_cols = list(set(demeaned.columns) - set(['scouting_team']))
demeaned.loc[:, numerical_cols] = demeaned.loc[:, numerical_cols] - demeaned.loc[:, numerical_cols].mean()
demeaned.head()
```

```python
A = demeaned.loc[:, numerical_cols].to_numpy().T
N = A.shape[1]
C = 1/N*(A@A.T)
r = np.linalg.eig(C)

variances = {}
for col, v in zip(numerical_cols, np.diag(C)):
    variances[col] = v

variances
```
